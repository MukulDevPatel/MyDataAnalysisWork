# Zomato Data Analysis Project

# Step 1 Importing libraries
1. Pandas is used for data manipulation and analysis
2. Numpy is used for numerical operations
3. Matplotlib.pyplot and Seaborn is used for data visualization.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Step 2 Create Data frame

```python
dataframe=pd.read_csv("Zomato data .csv")
print(dataframe)
```

```python
dataframe
```

# Convert the data type of column - rate

```python
def handleRate(value):
    value=str(value).split('/')
    value=value[0]
    return float(value)

dataframe['rate']=dataframe['rate'].apply(handleRate)
print(dataframe.head())
```

```python
dataframe.info()
```

# Type of resturant

```python
dataframe.head()
```

```python
sns.countplot(x=dataframe['listed_in(type)'])
plt.xlabel("type of resturant")
```

# Conclusion-majority of the resturant falls in dinning category

```python
dataframe.head()
```

```python
gouped_data=dataframe.groupby('listed_in(type)')['votes'].sum()
result=pd.DataFrame({'votes':gouped_data})
plt.plot(result,c='green',marker='o')
plt.xlabel("Type of resturant",c='red',size=20)
plt.ylabel('votes',c='red',size=20)
```

# conclusion-dinning resturants has recieved has maximum votes

```python
dataframe.head()
```

```python
plt.hist(dataframe['rate'],bins=5)
plt.title("Rating of distribution")
plt.show()
```

# conclusion-the majority resturants recieved ratings from 3.5 to 4.0

# Average order spending by couples

```python
dataframe.head()
```

```python
couple_data=dataframe['approx_cost(for two people)']
sns.countplot(x=couple_data,legend=False,palette='rainbow')
```

# conclusion-the majority of couples prefer resturants with an approximate cost of 300 rpees

# Which Mode recieve maximum ratings

```python
dataframe.head()
```

```python
plt.figure(figsize=(6,6))
sns.boxplot(x='online_order',y='rate',data=dataframe)
```

# conclusion-offline order recieved lower rating in comparison to online order

```python
dataframe.head()
```

```python
pivot_table=dataframe.pivot_table(index='listed_in(type)',columns='online_order',aggfunc='size',fill_value=0)
sns.heatmap(pivot_table,annot=True,cmap='YlGnBu',fmt='d')
plt.title("Heatmap")
plt.xlabel("Online Order")
plt.ylabel("Listed In type")
plt.show()
```

# Conclusion: Dinning resturants primarily accept offline orders,whereas cafes primarily recieve online orders. This cafes suggests that clients prefer to place orders in person at resturants, but prefer online ordering at cafes
