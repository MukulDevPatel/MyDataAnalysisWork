# Website Performance Analysis

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("data-export (1).csv")
```

```python
df
```

```python
df.columns=df.iloc[0]
df=df.drop(index=0).reset_index(drop=True)
df
```

```python
df
```

```python
df.columns=df.iloc[0]
df=df.drop(index=0).reset_index(drop=True)
df.columns=["channel group","DateHour","Users",	"Users Sessions","Engaged sessions",	"Average engagement time per session","Engaged sessions per user","Events per session","Engagement rate","Event count"]
```

```python
df
```

```python
df.info()
```

```python
df["DateHour"]=pd.to_datetime(df["DateHour"],format="%Y%m%d%H",errors='coerce')
```

```python
df
```

```python
numeric_cols=df.columns.drop(["channel group","DateHour"])
df[numeric_cols]=df[numeric_cols].apply(pd.to_numeric,errors='coerce')
df["Hour"]=df["DateHour"].dt.hour
```

```python
df
```

```python
df.info()
```

```python
df.describe()
```

# Sessions and User over time

```python
sns.set(style="whitegrid")
```

```python
plt.figure(figsize=(10,5))
df.groupby("DateHour")[["Users Sessions","Users"]].sum().plot(ax=plt.gca())
plt.title("Sessions and Users over time")
plt.xlabel("DateHour")
plt.ylabel("Count")
plt.show()
```

# Total users by channel

```python
plt.figure(figsize=(8,5))
sns.barplot(data=df,x="channel group",y="Users",estimator=np.sum,palette="viridis")
plt.title("Total Users by channel")
plt.xticks(rotation=45)
plt.show()
```

```python
df
```

# Average engagement time by channel

```python
plt.figure(figsize=(8,5))
sns.barplot(data=df,x="channel group",y="Average engagement time per session",estimator=np.mean,palette="magma")
plt.title("Average engagement time by Channel")
plt.xticks(rotation=45)
plt.show()
```

```python
df
```

# Engagement Rate Distribution by channel

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=df,x="channel group",y="Engagement rate",palette="coolwarm")
plt.title("Engagement Rate Distribution by Channel")
plt.xticks(rotation=45)
plt.show()
```

# Engaged vs Non engaged sessions

```python
df
```

```python
session_df=df.groupby("channel group")[["Users Sessions","Engaged sessions"]].sum().reset_index()
session_df["Non-Engaged"]=session_df["Users Sessions"]-session_df["Engaged sessions"]
session_df_melted=session_df.melt(id_vars="channel group",value_vars=["Engaged sessions","Non-Engaged"])

plt.figure(figsize=(8,5))
sns.barplot(data=session_df_melted,x="channel group",y="value",hue="variable")
plt.title("Engaged vs Non-Engaged Sessions")
plt.xticks(rotation=45)
plt.show()
```

# Traffic by Hour and Channel

```python
heatmap_data=df.groupby(["Hour","channel group"])["Users Sessions"].sum().unstack().fillna(0)
plt.figure(figsize=(12,6))
sns.heatmap(heatmap_data,cmap="YlGnBu",linewidths=.5,annot=True,fmt='.0f')
plt.title("Traffic by Hour and Channel")
plt.xlabel("Channel Group")
plt.ylabel("Hour of Day")
plt.show()
```

# Corelation Egagement rate and users sessions

```python
df_plot=df.groupby("DateHour")[["Engagement rate","Users Sessions"]].mean().reset_index()
plt.figure(figsize=(10,5))
plt.plot(df_plot["DateHour"],df_plot["Engagement rate"],label="Engagement rate",color="green")
plt.plot(df_plot["DateHour"],df_plot["Users Sessions"],label="Users Sessions",color="blue")
plt.title("Engagement Rate vs Sessions Over Time")
plt.xlabel("DateHour")
plt.legend()
plt.grid(True)
plt.show()
```
