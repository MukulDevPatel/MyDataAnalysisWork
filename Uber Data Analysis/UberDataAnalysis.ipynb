```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Data Pre-processing

```python
df['PURPOSE'].fillna("Not",inplace=True)
```

```python
df
```

```python
df['START_DATE']=pd.to_datetime(df['START_DATE'],errors='coerce')
df['END_DATE']=pd.to_datetime(df['END_DATE'],errors='coerce')
```

```python
df.info()
```

```python
from datetime import datetime
df['date']=pd.DatetimeIndex(df['START_DATE']).date
df['time']=pd.DatetimeIndex(df['END_DATE']).hour
```

```python
df
```

```python
df['day-night']=pd.cut(x=df['time'],bins=[0,10,15,19,24],labels=['Morning','Afternoon','Evening','Night'])
```

```python
df
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
df
```

# At what time do people book cabs the most from Uber

# In which months do people book Uber rides less frequently

```python
df
```

```python
df['Month']=pd.DatetimeIndex(df['START_DATE']).month
month_label={1.0:'Jan',2.0:'Feb',3.0:'Mar',4.0:'Apr',5.0:'May',6.0:'Jun',
            7.0:'July',8.0:'Aug',9.0:'Sep',10.0:'Oct',11.0:'Nov',12.0:'Dec'}
df['Month']=df.Month.map(month_label)
mon=df.Month.value_counts(sort=False)
```

```python
df
```

```python
dfp=pd.DataFrame({
    "Month":mon.values,
    "Value Count":df.groupby('Month',sort=False)['MILES'].max()
})
p=sns.lineplot(data=dfp)
p.set(xlabel="Month",ylabel="Value Count")  #Set the axis label for the plot
```

# On which days of the week do people book Uber rides the most

```python
df
```

```python
df['Day']=df.START_DATE.dt.weekday
day_label={
    0:'Mon',1:'Tue',2:'Wed',3:'Thu',4:'Fri',5:'Sat',6:'Sun'
}
df['Day']=df['Day'].map(day_label)
```

```python
df
```

```python
day_label=df.Day.value_counts()
sns.barplot(x=day_label.index,y=day_label)
plt.xlabel('Day')
plt.ylabel('Count')
plt.show()
```
